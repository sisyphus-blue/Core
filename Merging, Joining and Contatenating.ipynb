```python
import pandas as pd
w1=pd.read_csv("Restaurant - Week 1 Sales.csv")
w2=pd.read_csv("Restaurant - Week 2 Sales.csv")
c=pd.read_csv("Restaurant - Customers.csv")
f=pd.read_csv("Restaurant - Foods.csv")
```

```python
pd.concat([w1,w2],ignore_index=True)
```

```python
pd.concat([w1,w2],keys=["A","B"])
```

```python
w2.append(w1,ignore_index=True) # calls w1 on w2, so calls one 
```

```python
w1.merge(w2,how='inner',on='Customer ID',suffixes=['-A','-B'])
```

```python
w1.merge(w2,how="inner",on=["Customer ID", "Food ID"]) #common in w1 and w2. Inner join
```

```python
merged=w1.merge(w2,how="outer",on=["Customer ID","Food ID"],suffixes=["-Week 1","-Week 2"], indicator=True) 
# outer join
merged["_merge"].value_counts()
```

```python
merged[merged["_merge"].isin(['right_only','left_only'])] # no element from both
```

```python
w1.merge(f,how="left",on="Food ID",sort=True) # left join
```

```python
w2.merge(c,how="left",left_on="Customer ID",right_on="ID")
```

```python
f.set_index("Food ID",inplace=True)
```

```python
s1=w1.merge(c,how='left',left_on="Customer ID",right_index=True)
s1
```

```python
s2=w2.merge(f,how='left',left_on="Food ID",right_index=True)
s2
```

```python
w1.merge(w2,how='left',left_index=True,right_index=True,suffixes=[" Week 1"," Week 2"])
```
