```python
import pandas as pd
cg=pd.read_csv("chicago.csv").dropna(how="any")
```

```python
cg["Department"]=cg["Department"].astype('category')
```

```python
cg["Name"].str.lower()
```

```python
cg["Name"].str.title()
```

```python
cg["Department"]=cg["Department"].str.replace("MGMNT","MANAGEMENT")
```

```python
cg["Employee Annual Salary"]=cg["Employee Annual Salary"].str.replace("$","").astype('float')
```

```python
cg[cg["Position Title"].str.contains("WATER")] # CONTAINS
```

```python
cg[cg["Position Title"].str.startswith("POLICE")] # STARTSWITH
```

```python
cg[cg["Position Title"].str.endswith("IST")] # ENDSWITH
```
